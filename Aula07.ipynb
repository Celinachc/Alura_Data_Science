Importando pandas

```python
import pandas as pd
```

```python
import numpy as np
```

Preparando massa de dados

```python
from numpy.random import default_rng
```

```python
rng = default_rng()
```

```python
size = 100
base = {
'ids': rng.integers(low=15, high=1000, size=size),
'idades': rng.integers(low=18, high=120, size=size),
'sexo': rng.choice(['M','F'], size=size),
'estados_civis': rng.choice(['solteiro', 'casado', 'divorciado', 'viuvo'], size=size),
'sinistros': rng.integers(low=0, high=5, size=size),
'premios': rng.normal(loc=1000, scale=200, size=size)
}
```

Criando um DataFrame dos dados

```python
dataset = pd.DataFrame(base)
dataset.head()
```

Clientes com mais de 30 anos

```python
clientes_maior_30 = dataset[base['idades'] > 30]
```

```python
clientes_maior_30
```

Clientes sem sinistro utilizando query

```python
clientes_sem_sinistro = dataset.query('sinistros == 0')
```

```python
clientes_sem_sinistro
```

Exibindo idade e prêmio iterando o DataFrame



```python
for index, row in dataset.iterrows():
    print(row[1], row[-1])
```

Exibindo idade e prêmio selecionando colunas do DataFrame

```python
idades_premios = dataset[['idades','premios']]
```
